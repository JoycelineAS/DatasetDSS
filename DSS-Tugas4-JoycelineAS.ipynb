```python
import numpy as np
import pandas as pd
```

```python
#Nomor 1 (NOR)

# 1. Siapkan data bipolar (X1, X2)
X = np.array([
    [-1, -1],
    [-1, +1],
    [+1, -1],
    [+1, +1]
])

# 2. Siapkan target Y sesuai tabel kebenaran fungsi logika
#    Di sini: NOR (hanya +1 jika kedua input = -1)
Y = np.array([+1, -1, -1, -1])

# 3. Inisialisasi learning rate, bobot (w1, w2) dan bias
lr = 1
w = np.zeros(2)  # w = [w1, w2] awalnya [0.0, 0.0]
b = 0            # bias awalnya 0.0

# 4. Cetak header tabel untuk melacak proses training
print(f"{'Epoch':<5}{'X1':>5}{'X2':>5}{'Yin':>8}{'Out':>6}{'T':>8}"
      f"{'ΔW1':>6}{'ΔW2':>6}{'Δb':>6}{'W1':>8}{'W2':>6}{'b':>6}")

# 5. Mulai proses training hingga 10 epoch atau konvergen lebih cepat
for epoch in range(1, 11):
    changed = False  # indikator apakah ada perubahan bobot pada epoch ini

    # 5a. Loop untuk setiap sampel (x, t)
    for x, t in zip(X, Y):
        # Hitung net input Yin = w1*x1 + w2*x2 + b
        yin = np.dot(w, x) + b

        # Tentukan output perceptron: +1 jika Yin>=0, -1 jika Yin<0
        out = np.sign(yin)

        # Hitung error = target - output
        error = t - out

        # Hitung perubahan bobot dan bias
        dw = lr * error * x  # Δw = α * error * x
        db = lr * error      # Δb = α * error

        # Update bobot dan bias
        w += dw
        b += db

        # Tandai jika ada perubahan (error≠0)
        if error != 0:
            changed = True

        # Cetak kondisi saat ini dengan .0f agar angka bulat tampil tanpa .0
        print(f"{epoch:5d}{x[0]:5d}{x[1]:5d}"
              f"{yin:8.2f}{out:6.0f}{t:8.0f}"
              f"{dw[0]:6.0f}{dw[1]:6.0f}{db:6.0f}"
              f"{w[0]:8.0f}{w[1]:6.0f}{b:6.0f}")

    # Jika tidak ada satupun update bobot, training selesai lebih awal
    if not changed:
        print(f"\nTraining selesai di epoch ke-{epoch}.")
        break
```

```python
#Nomor 2

df = pd.read_csv('https://github.com/JoycelineAS/DatasetDSS/blob/main/tennis.csv')

df.head(15)
```

```python
import numpy as np

# 1) Siapkan data dan encode fitur
outlook_map = {'sunny': [1, 0, 0], 'overcast': [0, 1, 0], 'rainy': [0, 0, 1]}
temp_map    = {'hot'  : [1, 0, 0], 'mild'    : [0, 1, 0], 'cool': [0, 0, 1]}

X, y = [], []
for _, row in df.iterrows():
    feats = outlook_map[row['outlook']] + temp_map[row['temp']]
    feats.append(1 if row['humidity'] == 'high' else 0)
    feats.append(1 if row['windy'] else 0)
    X.append(feats)
    y.append(1 if row['play'] == 'yes' else 0)

X = np.array(X)
y = np.array(y)

# 2) Inisialisasi hyperparameter dan bobot
learning_rate = float(input("Masukkan alpha (learning rate): "))
threshold     = float(input("Masukkan theta (threshold): "))

weights = np.zeros(X.shape[1])
bias    = 0.0
epoch   = 1

# Cetak header training
print("\n=== Pelatihan Binary Perceptron ===")
print(f"{'Ep':<3}{'i':<3}{'Input':<30}{'Yin':>8}{'Out':>6}"
      f"{'Tgt':>6}{'Δb':>8}{'Δw':<30}{'W baru':<30}{'b baru':>8}")

# Loop utama training
while True:
    any_change = False
    for i, (x_vec, target) in enumerate(zip(X, y)):
        yin    = np.dot(weights, x_vec) + bias
        output = 1 if yin >= threshold else 0
        error  = target - output

        delta_b = learning_rate * error
        delta_w = learning_rate * error * x_vec

        bias    += delta_b
        weights += delta_w

        if error != 0:
            any_change = True

        # Convert list to str before formatting
        x_str       = str(x_vec.tolist())
        delta_w_str = str(delta_w.tolist())
        w_str       = str(weights.tolist())

        print(f"{epoch:<3}{i:<3}"
              f"{x_str:<30}"
              f"{yin:8.2f}{output:6d}{target:6d}"
              f"{delta_b:8.2f}"
              f"{delta_w_str:<30}"
              f"{w_str:<30}"
              f"{bias:8.2f}")

    if not any_change:
        print(f"\nTraining selesai di epoch ke-{epoch}.")
        break
    epoch += 1

# 3) Fungsi prediksi
def predict_playtennis(outlook, temp, humidity, windy, weights, bias, threshold):
    feats = outlook_map[outlook] + temp_map[temp]
    feats.append(1 if humidity == 'high' else 0)
    feats.append(1 if windy else 0)
    x_new = np.array(feats)
    yin   = np.dot(weights, x_new) + bias
    return "Play" if yin >= threshold else "Don’t Play"

# 4) Contoh penggunaan setelah training
result = predict_playtennis(
    outlook   = "sunny",
    temp      = "mild",
    humidity  = "normal",
    windy     = False,
    weights   = weights,
    bias      = bias,
    threshold = threshold
)
print("\nPrediction (binary):", result)
```

```python
# 1) Encode dataset ke representasi bipolar
outlook_labels = ['sunny', 'overcast', 'rainy']
temp_labels    = ['hot', 'mild', 'cool']

X_bipolar, y_bipolar = [], []
for _, row in df.iterrows():
    # satu-hot bipolar untuk outlook
    feats = [1 if row['outlook'] == lbl else -1 for lbl in outlook_labels]
    # satu-hot bipolar untuk temperature
    feats += [1 if row['temp']    == lbl else -1 for lbl in temp_labels]
    # fitur humidity & windy
    feats.append(1 if row['humidity'] == 'high' else -1)
    feats.append(1 if row['windy']      else -1)
    X_bipolar.append(feats)
    y_bipolar.append(1 if row['play'] == 'yes' else -1)

X_bipolar = np.array(X_bipolar)
y_bipolar = np.array(y_bipolar)

# 2) Inisialisasi bobot, bias, dan hyperparameter
alpha     = float(input("Masukkan alpha (bipolar): "))
theta     = float(input("Masukkan theta (bipolar): "))
weights   = np.zeros(X_bipolar.shape[1])
bias      = 0.0
epoch     = 1

print("\n=== Pelatihan Bipolar Perceptron ===")
print(f"{'Ep':<3}{'i':<3}{'Input':<30}{'Yin':>8}{'Out':>6}"
      f"{'Tgt':>6}{'Δb':>8}{'Δw':<30}{'W baru':<30}{'b baru':>8}")

# 3) Training loop
while True:
    any_change = False
    for i, (x_vec, target) in enumerate(zip(X_bipolar, y_bipolar)):
        yin    = np.dot(weights, x_vec) + bias
        out    = 1 if yin >= theta else -1
        error  = target - out

        delta_b = alpha * error
        delta_w = alpha * error * x_vec

        bias    += delta_b
        weights += delta_w

        if error != 0:
            any_change = True

        # convert list ke string agar f-string tidak error
        x_str    = str(x_vec.tolist())
        dw_str   = str(delta_w.tolist())
        w_str    = str(weights.tolist())

        print(f"{epoch:<3}{i:<3}"
              f"{x_str:<30}"
              f"{yin:8.2f}{out:6d}{target:6d}"
              f"{delta_b:8.2f}{dw_str:<30}{w_str:<30}{bias:8.2f}")

    if not any_change:
        print(f"\nTraining selesai pada epoch ke-{epoch}.")
        break
    epoch += 1

# 4) Fungsi prediksi yang mengembalikan string
def predict_playtennis_bipolar(outlook, temp, humidity, windy, weights, bias, theta):
    feats = [1 if outlook == lbl else -1 for lbl in outlook_labels]
    feats += [1 if temp    == lbl else -1 for lbl in temp_labels]
    feats.append(1 if humidity == 'high' else -1)
    feats.append(1 if windy      else -1)
    x_new = np.array(feats)
    yin   = np.dot(weights, x_new) + bias
    return "Play" if yin >= theta else "Don’t Play"

# 5) Contoh pemakaian setelah training
result = predict_playtennis_bipolar(
    outlook='rainy',
    temp='cool',
    humidity='high',
    windy=True,
    weights=weights,
    bias=bias,
    theta=theta
)
print("\nPrediction (bipolar):", result)
```
